```python
#calculate total no. of statewise population
data.groupby('State_name').Population.sum()
```

```python
#search for the data
data['Education_status']=data['Secondary_Education']>100000
```

```python
#print the data with condition
data[data['Education_status'].isin(['GOOD'])]
```

```python
#to find data according to religion sum of population
data.columns
data.groupby('Population')['Hindus', 'Muslims', 'Christians', 'Sikhs', 'Buddhists', 'Jains'].sum()
```

```python
#sort values according to hindus
data.groupby('Population')['Hindus', 'Muslims', 'Christians', 'Sikhs', 'Buddhists', 'Jains'].sum().sort_values(by = 'Hindus')
```

```python
#TO COUNT NO OF MALE WORKERS OF PARTICULAR STATE
#to find ratio of male is to female workers
(data[data.State_name == 'JHARKHAND']['Male_Workers'].sum())/(data[data.State_name == 'JHARKHAND']['Female_Workers'].sum())
```
